```python
import matplotlib as mpt
import matplotlib.pyplot as plt 
import pandas as pd 
import scienceplots
import numpy as np
import os
plt.style.use(['science', 'notebook', 'grid'])
df = pd.read_csv("physlab - Sheet1 (1).csv")
df['Avg'].values


x = np.arange(0, 100, 1)
x_data = np.arange(25, 75, 10)
y_data = df['Avg'].values
g = 9.8
V = 3.41175
err = df['STD'].values


def prj(x):
    return V**2/g * np.sin(np.radians(2*x))

d = prj(x)
plt.figure(figsize=(14,8))
plt.errorbar(x_data, y_data, fmt = 'ro', yerr = 0.1446518303, capsize=5, capthick=1, ecolor = 'black')
plt.plot(x, d, 'o', color='purple', ms=2, label = 'Theoretical distribution')
plt.plot(x_data, y_data, 'o', color='red', ms=10, label = 'Experimental Data')
plt.ylabel('Distance (m)')
plt.xlabel('Angle (degrees)')
plt.legend(loc='upper right', fontsize=12)
plt.savefig('physlab.png', dpi = 1000)
plt.show()


    



```
